## Polar Axis

For a demonstration, see @fig-polar

```python
#| label: fig-polar
#| fig-cap: "Let see"

import numpy as np
import matplotlib.pyplot as plt

x = np.arange(1, 2, 100)
y = x ** 2
plt.plot(x, y)
plt.show()
```
